# What factors are driving human population density?

Let's first get some data on human population density:

```r
library(raster)

human_pop_file='/Users/tobias/Desktop/advanced_spatial_r/data/human_pop_density/ascii/gpw_v4_population_count_rev11_2020_2pt5_min.asc'
human_pop_data = raster(human_pop_file)
```

```r
human_pop_data
```

Load the shape of the world map:

```r
library(rworldmap)
world_map <- getMap(resolution = "li")
```

Now plot the raw human population data on the world map

```r
plot(world_map)
plot(human_pop_data,add=T)
```

Displaying the raw values makes it difficult to get an impression of the data, since there are some areas with very high population density that stretch the color-scale so that makes it impossible to see differences in the data in most parts of the world. A common way to fix this is to log-transform your data, which compresses the range of the data, making small differences more visible.

```r
plot(world_map)
plot(log(human_pop_data),add=T)
```
